# plant1_train 정제

```python
with open('plant1_train.txt') as data:
    lines = data.readlines()
```

```python
len(lines)
```

```python
value_ls = []
for i in range(len(lines)):

    a = lines[i].split(',')
    
    value_ls.append(a)
        
```

```python
plant1_train = pd.DataFrame(value_ls[1:], columns = value_ls[0])
```

```python
plant1_train.drop(columns = '"Index"',inplace = True)
```

```python
columns = []
for i in value_ls[0]:
    columns.append(re.sub('[\"\n]','',i))
```

```python
plant1_train.columns = columns[1:]
```

```python
col = []
for i in range(len(plant1_train)):
    col0 = re.sub('[\n]','',plant1_train['plant1_train.cond_loc3'].iloc[i])
    col.append(col0)
```

```python
plant1_train['plant1_train.cond_loc3'] = col
```

```python
for i in plant1_train.columns:
    plant1_train[i] = plant1_train[i].replace('NA',np.nan)
```

```python
for i in plant1_train.columns[1:]:
    plant1_train[i] = plant1_train[i].astype('float64')
```
